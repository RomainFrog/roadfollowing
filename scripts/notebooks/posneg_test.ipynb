```python
import torch
import torchvision.transforms as transforms
from PIL import Image
from torchvision import models
import torch.nn as nn
import warnings
import os

warnings.filterwarnings("ignore", category=UserWarning)
# Define the transforms to be applied to the input images
transform = transforms.Compose([
    transforms.Resize((244, 244)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])])

# Create an instance of the ResNet18 model
model = models.resnet18(pretrained=True)
model.fc = nn.Linear(512, 2)
model.load_state_dict(torch.load('models/binary_classifier.pth'))

# Load the saved state dictionary

# Set the model to evaluation mode
model.eval()

for filename in os.listdir("neg_augmented"):
    # Load a new image and apply the transforms
    image = Image.open('neg_augmented/' + filename)
    # image = Image.open('posneg_dataset/neg/d88e51f0-cb2d-11ed-8e12-fc4482a56b58.jpg')
    image = transform(image)

    # Add a batch dimension to the image
    image = image.unsqueeze(0)

    # Make a prediction on the image
    with torch.no_grad():
        output = model(image)

    # Get the predicted class label
    pred = torch.sign(output[0][1]).item()

    # Print the predicted label
    if pred == 1:
        print('Positive')
        print(filename)
    else:
        print('Negative')
```
